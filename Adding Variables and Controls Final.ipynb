```r
# Determine if there was a senate election in a given state in a given year
senate_values <- numeric(nrow(final))

# Iterate over each year from 2000 to 2020
for (year in seq(2000, 2020, by = 2)) {
  # Get the Senate vector for the current year
  senate_vector <- get(paste0("senate_", year))
  # Iterate over each row of the final data frame
  for (i in 1:nrow(final)) {
    if (final$STATE[i] %in% senate_vector & final$Year[i]==year) {
      # If present, set the corresponding element in senate_values to 1
      senate_values[i] <- 1
    }
  }
}

# Assign the senate_values vector to the senate column in the final data frame
final$senate <- senate_values
```
