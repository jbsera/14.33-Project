# 2020 Data

## 2020 Race and Sex Data

```r
#Turning xl file into csv
write.csv(read_excel("C:\\Users\\joyse\\Desktop\\14.33\\Data\\table04b_2020.xlsx"), "C:\\Users\\joyse\\Desktop\\14.33\\Data\\table04b_2020.csv", row.names = FALSE)
```

```r
#Skipping comments in the original excel file
race_2020 <- read.csv("C:\\Users\\joyse\\Desktop\\14.33\\Data\\table04b_2020.csv",  skip = 5)
race_2020 <- race_2020[-(573:nrow(race_2020)), ]
```

```r
#Checking that I skipped correctly
tail(race_2020, 20)
```

```r
#Naming the columns 
col_names <- c("STATE", "Sex, Race, and Hispanic-Origin", "Total population", "Total citizen population", "Total registered", 
               "Percent registered (Total)", "Margin of error for percent registered total", "Percent registered (Citizen)",
               "Margin of error for percent registered citizen", "Total voted", "Percent voted (Total)", "Margin of error percent voted total", 
               "Percent voted (Citizen)", "Margin of error percent voted citizen")
colnames(race_2020) <- col_names
```

```r
#Creating the state titles and the year variable
race_2020$STATE <- na.locf(race_2020$STATE)
race_2020$Year <- 2020
```

```r
#Changing the '-' entries to 0, as they signify
race_2020[race_2020 == '-'] <- 0
```

```r
#Converting columns to numerics
race_2020$`Total registered` <- as.numeric(race_2020$`Total registered`)
race_2020$`Total citizen population` <- as.numeric(race_2020$`Total citizen population`)
race_2020$`Total voted` <- as.numeric(race_2020$`Total voted`)
```

```r
head(race_2020)
```

```r
#Recalculating the percents to get rid of the "B's"
race_2020$`Percent registered (Total)` <- ifelse(race_2020$`Total population` != 0,
                                                 100*race_2020$`Total registered` / race_2020$`Total population`,
                                                 0)

race_2020$`Percent registered (Citizen)` <- ifelse(race_2020$`Total citizen population` != 0,
                                                 100*race_2020$`Total registered` / race_2020$`Total citizen population`,
                                                 0)

race_2020$`Percent voted (Total)` <- ifelse(race_2020$`Total population` != 0,
                                                 100*race_2020$`Total voted` / race_2020$`Total population`,
                                                 0)

race_2020$`Percent voted (Citizen)` <- ifelse(race_2020$`Total citizen population` != 0,
                                                 100*race_2020$`Total voted` / race_2020$`Total citizen population`,
                                                 0)



```

```r
head(race_2020)
```

```r
#Pivoting to the desired format
race_2020_final <- pivot_wider(race_2020, 
                          names_from = `Sex, Race, and Hispanic-Origin`, 
                          values_from = c("Total population", "Total citizen population", 
                                          "Total registered", "Percent registered (Total)", 
                                          "Margin of error for percent registered total", 
                                          "Percent registered (Citizen)", 
                                          "Margin of error for percent registered citizen", 
                                          "Total voted", "Percent voted (Total)", 
                                          "Margin of error percent voted total", 
                                          "Percent voted (Citizen)", 
                                          "Margin of error percent voted citizen"))
```

```r
head(race_2020_final)
```

```r
write.csv(race_2020_final, "C:\\Users\\joyse\\Desktop\\14.33\\Data\\table04b_2020_final.csv", row.names = FALSE)
```
