#### Some general notes
1) Merging is fairly straightforward
2) Sometimes the age group labels for the different years don't coincide. I didn't worry about this since it wasn't possible to overlap these groups and since age is not my primary metric of interest
3) Sometimes the race categories differ (2002 and 2000). In this case, I grouped Asian with Asianaloneorincombination, Black with Blackaloneorincombination and etc.
